# preforming machine learning 

```python
import pandas as pd
import numpy as np
import sklearn 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn .linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
dsf["Name"]=le.fit_transform(dsf["Name"])
```

```python
dsf["Kilometer_Driven"]=le.fit_transform(dsf["Kilometer_Driven"])
```

```python
dsf["Address"]=le.fit_transform(dsf["Address"])
```

```python
dsf["Price"]=le.fit_transform(dsf["Price"])
```

```python
#some example of univariate analysis
```

```python
sns.violinplot(x="Name",data=dsf)
plt.show()
```

```python
sns.violinplot(x="Kilometer_Driven",data=dsf)
plt.show()
```

```python
sns.violinplot(x="Address",data=dsf)
plt.show()
```

```python
dsf['Kilometer_Driven'].plot.box()
```

```python
sns.scatterplot(data=dsf,x='Name',y='Kilometer_Driven')
```

```python
x=dsf.drop("Price",axis=1)
y=dsf['Price']
```

```python
x
```

```python
y
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=.22,random_state=137)
```

```python
train_x.shape
```

```python
test_x.shape
```

```python
train_y.shape
```

```python
test_y.shape
```

```python
lm=LinearRegression()
```

```python
lm.fit(train_x,train_y)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
lm.score(train_x,train_y)
```

```python
#predict the value 
pred=lm.predict(test_x)
print("predicted result price:",pred)
print("actual price",test_y)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
print("error:")
print('mean absolute error:',mean_absolute_error(test_y,pred))
print('mean squared error:',mean_squared_error(test_y,pred))
```

```python
from sklearn.metrics import r2_score
print(r2_score(test_y,pred))
```

```python
from sklearn.linear_model import Lasso,Ridge
ls=Lasso(alpha=0.0001)
ls.fit(train_x,train_y)
ls.score(train_x,train_y)
```

```python
ls.coef_
```

```python
rd=Ridge(alpha=0.0001)
rd.fit(train_x,train_y)
rd.score(train_x,train_y)
```

```python
rd.coef_
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=100,random_state=137)
rf.fit(train_x,train_y)
predrf=rf.predict(test_x)
```

```python
rf.score(train_x,train_y)
```
